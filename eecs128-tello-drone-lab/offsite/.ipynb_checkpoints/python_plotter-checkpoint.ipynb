```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open("./statedata.txt") as f:
    data = f.read().split("\n")[3:]
    ts = []
    cs = []
    ys = []
    for line in data:
        line = line.split(",")
        try:
            time = float(line[1].strip())
        except Exception:
            continue
        control_yaw = float(line[6].strip())
        yaw = float(line[9].strip())
        ts.append(time)
        cs.append(control_yaw)
        ys.append(yaw)
    time = ts
    control_yaw = cs
    yaw = ys
```

```python
import numpy as np
ref = lambda ptime: (2 * 180) / np.pi * np.arcsin(np.sin(2*np.pi * ptime / 20))

plt.figure(figsize=(12, 9))
plt.plot(time, yaw)
plt.plot(time, [ref(t) for t in time])
#plt.plot(time, control_yaw)
```
